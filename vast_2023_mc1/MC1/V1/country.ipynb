```python
# 只和一个国家有关系
only_one=[]
for i in source_flow.keys():
    src_cty=source_flow[i]
    for j in src_cty:
        tgt_cty=source_flow[i][j]
        if (i!=j) and (len(src_cty)==1) and (j!='unknown'):
            only_one.append((i,j))
print(len(only_one),only_one)
```

```python
suspicious_country_nodes=[]
# 国家只有一个东西但是贸易往来比较多
for i in source_flow.keys():
    src_cty=source_flow[i]
    if (src_cty['country_num']==1):
        if (len(src_cty)>2) and (i not in src_cty):
            suspicious_country_nodes.append(i)
        else:
            for j in src_cty:
                tgt_cty=source_flow[i][j]
                if tgt_cty>10:
                    suspicious_country_nodes.append(i)
suspicious_country_nodes
```

### 往来关系

```python
# 去掉family relationship看
merged_3ship=merged_info[merged_info['link_type']!='family_relationship']
exp_country_links=merged_3ship.groupby('source_country')
print(len(exp_country_links))
source_flow_3={}
for i in exp_country_links:
    exp_cty,exp_df=i
    source_flow_3[exp_cty]=(exp_df['target_country'].value_counts().to_dict())
    source_flow_3[exp_cty]['country_num']=len(nodes_df[nodes_df['country']==exp_cty])
type(source_flow_3['Ariuzima'])
```

```python
avg_between_coutries=[]
sus_country=[]
sum_all=0
for i in source_flow_3.keys():
    src_cty=source_flow_3[i]
    sum_cty=0
    flag=0
   
    other_countries=0
    for j in src_cty:
        tgt_cty=source_flow_3[i][j]
        # countrynum都在最后反正
        if j=='country_num':
            break
        if i!=j:
            sum_cty+=tgt_cty
        #如果有和自己国家往来的情况: 
        else:
            flag=1
    # 因为有一个country_num，所以要减去1
    if len(src_cty)!=1:
        other_countries=len(src_cty)-1
    # 只和unknown，就只有country num，就是0
    else:
        other_countries=0
    # 不算自己国家内部的交易数量
    if flag==1:
        other_countries-=1
    # 如果只和自己国家交易：
    if other_countries==0:
        sum_cty=0
    else:
        sum_cty=sum_cty/other_countries
    avg_between_coutries.append((i,sum_cty))
    sum_all+=sum_cty
sum_all/=len(source_flow_3.keys())
print(avg_between_coutries)
# 把0的去掉
no_zero=[i for i in avg_between_coutries if i[1]!=0]
mean_all=np.mean([i[1] for i in no_zero])
print(mean_all)
larger=np.where(np.array([i[1] for i in avg_between_coutries])>=mean_all)
print(sum_all)
for j in larger[0]:
    # print(j)
    # print(avg_between_coutries[int(j)])
    # 和每个国家的平均来往数量大于所有的平均，并且本身属于该国的成员很少
    if (avg_between_coutries[int(j)][1] > sum_all) and (len(countries_dict[avg_between_coutries[j][0]])<10):
        print(avg_between_coutries[int(j)])
        sus_country.append(avg_between_coutries[int(j)])
```
