```python
import json
import networkx as nx
import numpy as np
import pandas as pd

with open('./MC1.json', 'r') as file:
    json_data = json.load(file)
nodes = json_data['nodes']
links = json_data['links']

G = nx.DiGraph()
edges = []
edges1 = []
edges2 = []
for link in links:
    edges.append((link['source'], link['target']))
G.add_edges_from(edges)

for component in nx.strongly_connected_components(G):
    component0 = list(component)
    for i in range(len(component0)):
        component0[i]=str(component0[i])

    if 'Mar de la Vida OJSC' in component0 or '979893388' in component0 or 'Oceanfront Oasis Inc Carriers' in component0 or '8327' in component0:
        if len(edges1) == 0:
            subgraph = G.subgraph(component)
            edges1 = list(subgraph.edges())
        else:
            subgraph = G.subgraph(component)
            edges2 = list(subgraph.edges())
            break

for i in range(len(edges1)):
    edges1[i]=(str(edges1[i][0]), str(edges1[i][1]))

print(len(edges1))
print(edges1)

for i in range(len(edges2)):
    edges2[i]=(str(edges2[i][0]), str(edges2[i][1]))

print(len(edges2))
print(edges2)


df=pd.read_csv('./merged_info.csv')
# "Mar de la Vida OJSC",
# "979893388",
# "Oceanfront Oasis Inc Carriers",
# "8327"
```

# Ownership

```python
arr1 = np.zeros((9,9))
o_table=pd.DataFrame(arr1, columns=['company_s', 'event_s', 'location_s', 'movement_s', 'organization_s', 'person_s', 'political_organization_s', 'vessel_s', 'unknown_s'],index=['company_t', 'event_t', 'location_t', 'movement_t', 'organization_t', 'person_t', 'political_organization_t', 'vessel_t', 'unknown_t'])

o_table_4points=pd.DataFrame(columns=['company_s', 'event_s', 'location_s', 'movement_s', 'organization_s', 'person_s', 'political_organization_s', 'vessel_s', 'unknown_s'],index=['company_t', 'event_t', 'location_t', 'movement_t', 'organization_t', 'person_t', 'political_organization_t', 'vessel_t', 'unknown_t'])
o_table_4points['organization_s'] = 0
o_table_4points['unknown_s'] = 0
o_table_4points.iloc[4] = 0
o_table_4points.iloc[8] = 0

o_cnt = 0
```

```python
collect_4points = pd.DataFrame()
for row in df.iterrows():
    # print(row[1][1],row[1][7],row[1][9])
    st=(row[1][4],row[1][5])
    # print(type(row[1][7]))
    if row[1][1] == 'ownership' and (st in edges1):
        o_cnt = o_cnt + 1
        o_table[row[1][7]+'_s'][row[1][9]+'_t'] = o_table[row[1][7]+'_s'][row[1][9]+'_t'] + 1

        if (row[1][4] == 'Mar de la Vida OJSC') or (row[1][5] == 'Mar de la Vida OJSC') or (row[1][4] == '979893388') or (row[1][5] == '979893388') or (row[1][4] == 'Oceanfront Oasis Inc Carriers') or (row[1][5] == 'Oceanfront Oasis Inc Carriers') or (row[1][4] == '8327') or (row[1][5] == '8327'):
            collect_4points = collect_4points.append(row[1])
            o_table_4points[row[1][7]+'_s'][row[1][9]+'_t'] = o_table_4points[row[1][7]+'_s'][row[1][9]+'_t'] + 1

collect_4points
```
